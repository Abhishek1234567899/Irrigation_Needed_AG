```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for the curve (quadratic function)
x = np.linspace(-10, 10, 100)  # 100 points from -10 to 10
a, b, c = 1, 0, 0  # Coefficients for the quadratic function y = ax^2 + bx + c
y = a * x**2 + b * x + c  # Quadratic equation

# Create the plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(x, y, label='Quadratic Curve: $y = x^2$', color='orange')  # Plot the curve

# Highlight the vertex point
vertex_x = -b / (2 * a)  # x-coordinate of the vertex
vertex_y = c - (b**2) / (4 * a)  # y-coordinate of the vertex
plt.scatter(vertex_x, vertex_y, color='red', zorder=5)  # Mark the vertex
plt.text(vertex_x, vertex_y, f' Vertex ({vertex_x}, {vertex_y})', fontsize=10, ha='right', color='red')

# Add annotations for important points
plt.annotate('Increasing', xy=(5, 25), xytext=(6, 40), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Decreasing', xy=(-5, 25), xytext=(-6, 40), arrowprops=dict(facecolor='black', shrink=0.05))

# Customize the plot
plt.title('Curve-wise Visualization of Quadratic Function')  # Title of the plot
plt.xlabel('X-axis')  # Label for X-axis
plt.ylabel('Y-axis')  # Label for Y-axis
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # Horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5, ls='--')  # Vertical line at x=0
plt.grid(color='gray', linestyle='--', linewidth=0.5)  # Add grid
plt.legend()  # Show legend
plt.show()  # Display the plot
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt

# Example data (replace this with your actual dataset)
data = {
    'Growth_Stage': ['Reproductive', 'Ripening', 'Seedling', 'Vegetative'],
    'Mean_10cm': [24.464780, 24.738384, 23.157918, 24.892128],
    'Std_10cm': [4.0, 5.0, 3.5, 4.2],  # Standard deviations for each growth stage at 10cm
    'Mean_30cm': [32.678912, 32.657581, 30.959837, 31.388921],
    'Std_30cm': [5.0, 5.5, 4.0, 4.5],  # Standard deviations for each growth stage at 30cm
    'Mean_50cm': [40.166334, 37.002323, 44.189566, 40.200088],
    'Std_50cm': [6.0, 5.8, 6.5, 5.6]   # Standard deviations for each growth stage at 50cm
}

df = pd.DataFrame(data)

# Define function to plot normal distribution curves
def plot_normal_distribution(depth, growth_stage):
    mean = df.loc[df['Growth_Stage'] == growth_stage, f'Mean_{depth}'].values[0]
    std = df.loc[df['Growth_Stage'] == growth_stage, f'Std_{depth}'].values[0]
    
    # Generate synthetic data based on mean and std
    x = np.linspace(mean - 4*std, mean + 4*std, 100)
    y = stats.norm.pdf(x, mean, std)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'Normal Distribution (mean={mean}, std={std})')
    plt.title(f'Normal Distribution of Soil Moisture at {depth} Depth during {growth_stage}')
    plt.xlabel('Soil Moisture (%)')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Plot the normal distribution curves for each depth during a specific growth stage
plot_normal_distribution('10cm', 'Vegetative')
plot_normal_distribution('30cm', 'Vegetative')
plot_normal_distribution('50cm', 'Vegetative')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data (replace this with your actual dataset)
data = {
    'Growth_Stage': ['Reproductive', 'Ripening', 'Seedling', 'Vegetative'],
    'Percentile25_10cm': [18.212362, 17.017650, 15.624155, 15.629097],
    'Median_10cm': [22.913383, 25.469553, 22.097277, 25.228674],
    'Percentile75_10cm': [32.825629, 32.784297, 28.275053, 34.002437],
    'Percentile25_30cm': [21.358961, 24.418320, 24.047442, 23.024843],
    'Median_30cm': [35.328523, 34.548242, 28.668211, 32.527052],
    'Percentile75_30cm': [43.275994, 38.916362, 38.787936, 38.667251],
    'Percentile25_50cm': [30.924307, 25.111523, 34.486322, 31.976189],
    'Median_50cm': [41.439795, 38.952912, 46.220740, 41.297758],
    'Percentile75_50cm': [47.775437, 47.155238, 50.282539, 47.818901]
}

df = pd.DataFrame(data)

# Define function to plot box plot for each depth
def plot_boxplot(depth):
    data_for_boxplot = []
    for stage in df['Growth_Stage']:
        lower_quartile = df.loc[df['Growth_Stage'] == stage, f'Percentile25_{depth}'].values[0]
        median = df.loc[df['Growth_Stage'] == stage, f'Median_{depth}'].values[0]
        upper_quartile = df.loc[df['Growth_Stage'] == stage, f'Percentile75_{depth}'].values[0]
        data_for_boxplot.append([lower_quartile, median, upper_quartile])
    
    plt.figure(figsize=(10, 6))
    plt.boxplot(data_for_boxplot, positions=range(1, len(df['Growth_Stage']) + 1), patch_artist=True)
    plt.xticks(range(1, len(df['Growth_Stage']) + 1), df['Growth_Stage'])
    plt.title(f'Soil Moisture Distribution at {depth} Depth')
    plt.xlabel('Growth Stage')
    plt.ylabel('Soil Moisture (%)')
    plt.grid(True)
    plt.show()

# Plot the box plot for each depth
plot_boxplot('10cm')
plot_boxplot('30cm')
plot_boxplot('50cm')
```
